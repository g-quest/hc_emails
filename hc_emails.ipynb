### Problem 1  
Create a list of all emails that mention Syria. How many are there?

```python
df = df.fillna('missing data') #fill missing data
```

```python
containsSyria = df[df["ExtractedBodyText"].str.contains("Syria", case = False)] #filter to messages containing 'Syria'
```

### Problem 4 
Write a function that returns the *n* most frequently emailed (from Clinton) addresses. (So if I want the 10 most commonly emailed addresses, I input 10 as a parameter to the function).

```python
def most_freq(n):
    sort_email = hilDF["MetadataTo"].value_counts() #group by email addresses then count occurrences
    top_n = sort_email.head(n) #take the top n most frequently emailed addresses
    print(top_n)
```

```python
#Call the function with the desired number of most frequent emails. For example:

#For the 5 most commonly emailed addresses:
most_freq(5)
```

```python
#For the 10 most commonly emailed addresses:
most_freq(10)
```

### Problem 5  
Find all the world leaders that Hillary may have had contact with. Find the names of all the world leaders in the emails, including their titles.

*Hint: what are the potential titles of world leaders? How would you distinguish them from unrelated information? For example, “PM” for prime minister versus “PM” for time of day. You will want to use REGEX*


```python
import re

#regex finds all world leader titles and associated name
findLeaders = []
for email in hilDF['ExtractedBodyText']:
    regex = re.findall(r'President [A-Z]{1}[a-z]*|PM [A-Z]{1}[a-z]*|Prime Minister [A-Z]{1}[a-z]*|FM [A-Z]{1}[a-z]*|First Minister [A-Z]{1}[a-z]*', email, re.MULTILINE)
    if len(regex) > 0:
        findLeaders.append(regex)

#clean out list and remove duplicates
cleanLeaders = set([val for sublist in findLeaders for val in sublist])

#replace acronyms and sort
leaders = []
for name in cleanLeaders:
    name = name.replace('FM', 'First Minister')
    name = name.replace('PM', 'Prime Minister')
    leaders.append(name)
worldLeaders = sorted(leaders)

#show world leaders
for each in worldLeaders: 
    print(each)
```
